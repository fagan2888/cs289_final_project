```python
#Create one panel that combines all of the error dataframes produced
combined_err_panel = pd.concat([err_panel, err_panel_1200, err_panel_800, err_panel_400], ignore_index=True)
#Create a dictionary of error panels, where the keys are the sample size used to produce the dataframes in that panel
#or the word "all" for the panel which combines all the error dataframes
error_panels = dict(zip([1247, 1200, 800, 400, "all"],
                        [err_panel, err_panel_1200, err_panel_800, err_panel_400, combined_err_panel]))
#Get a list of the variables that can be seen in any error dataframe
variables = err_panel[0].index.tolist()
#Get a list of the callifier labels used for the columns in each error dataframe
classifier_names = err_panel[0].columns.tolist()
#initialize a dictionary to contain all of the size, then variable organized error dataframes
var_panels = {}
#Iterate over each sample size, including "all" which joins all trials regardless of sample size
for size in error_panels:
    panel = error_panels[size] #Isolate the panel of error dataframes which corresponds to that sample size
    panel_keys = panel.keys() #Get the keys used in that panel of dataframes
    var_panels[size] = {} #Initialize a dictionary of variable dataframes for this sample size
    for variable in variables: #Iterate through all the diagnostic variables
        data_dict = {} #Initialize a dictionary to hold the data for this diagostic variable from the various trials.
        for classifier in classifier_names: #Iterate through each of the classifiers
            key = "{} {}".format(classifier, variable) #Create a name for the column in the variable's dataframe
             #Get the values of this variable from each dataframe in the error panel
            data_dict[key] = [panel[x].at[variable, classifier] for x in panel_keys]
        #Create the dataframe of scores on this variable for the given sample
        var_panels[size][variable] = pd.DataFrame(data_dict)
```

```python
#initialize a variable for the path to the folder where I want to store the figures for my paper
figure_folder = "/home/timothy/Spring 2014 UCB/CE265-Traffic Safety/finalProject/figures/multiple_trials"
for size in var_panels.keys(): #Iterate over all the training set sizes
    for variable in variables[:-2]: #Iterate over all the variables for which I have single scores from each classifier
        df = var_panels[size][variable] #Isolate the appropriate error dataframe
        data = [df[x] for x in df.columns] #Get the data from the various classifiers
        plt.boxplot(data) #Create a boxplot witht he classifiers on the x-axis and scores on the y-axis
        plt.xticks([1,2,3], [x[:-(len(variable) + 1)] for x in df.columns]) #Label the positions on the x-axis
        plt.xlabel("Classifier Type") #Label the x-axis as a whole
        plt.ylabel("{}".format(variable)) #Label the y-axis
        if isinstance(size, str): #Create a title for the figure based on the training set size
            plot_title = "{} of each Classifier Over all Trials Examples".format(variable)
        else:
            plot_title = "{} of each Classifier Over {} Trials of {} Examples".format(variable, len(df), size) + "\n"
        plt.title(plot_title) #Create a tittle on the figure
        plt.savefig(os.path.join(figure_folder, "{}_{}.png".format(variable, size))) #Save the figure
        plt.show() #Display the figure inline
```
